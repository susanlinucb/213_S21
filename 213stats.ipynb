Notice that the way to look at the "head" of a dataframe is different in R...

```r
df = read.csv('./213_data_normalized.csv')
head(df)
```

R has a nice subsetting function!

```r
subset(df,df$measurement=='oaf')  # The $ symbol accesses variables
```

```r
subset(df,df$value > 0)  # The $ symbol accesses variables
```

R's generic `plot` function is a "smart" function, meaning what it displays depends on the variables you give it. By default, if you plot two numeric values against each other, you'll get a scatterplot:

```r
plot(value~time,data=df)
```

But a scatterplot is a pretty terrible way to visualize this data, beacuse the time values 20/50/80 are not really a continuous variable. We can force the dataframe to treat that variable as a discrete variable instead. Now, the plot function will give us a boxplot instead!

```r
df$time = as.factor(df$time)
plot(value~time,data=df)
```

```r
head(df)
```

What if we give it two discrete variables?

```r
plot(Height2~Height1,data=df)
```

What if we want to make more complex plots? A simple package is the `lattice` package. A more complex package is `ggplot`, but (in my opinion) you are better off using `seaborn` in python if you need `ggplot`'s complexity.

```r
library(lattice)
bwplot(value~time|measurement,data=df, ylim=c(-2,3))
bwplot(value~Nasal|time*measurement,data=df, ylim=c(-2,3))
```

Ok but that's not what we moved to R for... we want to try to do some stats!

```r
summary(aov(value~Nasal, data=subset(df,df$measurement=='naf' & df$time==80)))
```

```r
summary(aov(value~Nasal+Error(df$Repetition), data=subset(df,df$measurement=='naf' & df$time==80)))
```

Let's do something with the nasal context first, just to confirm that things "look good". In the context of a nasal coda, we should expect nasal airflow (measurement=naf) to be higher than when the vowels are not in the context of a nasal coda. For simplicity's sake, let's look only at the data at the ends of the vowel (time=80)

```r
summary(lmer(value~Nasal+(1+Nasal|Speaker)+(1+Nasal|Word),
             data=subset(df,df$measurement=='naf' & df$time==80)))
```

But you might be concerned that there is no p-value. If you require p-values, you can use the library lmerTest or a different method.

```r
install.packages('lmerTest')
library(lmerTest)
```

```r
summary(lmer(value~Nasal+(1+Nasal|Speaker)+(1+Nasal|Word),
             data=subset(df,df$measurement=='naf' & df$time==80)))
```
